### Pie charts are commonly used for visualizing categorical data, particularly when you want to represent the proportion of each category relative to the whole.

```python
y=np.array([10,135,25,25,19])
plt.pie(y)
plt.show()
```

```python
#with labels
ny=np.array([20,34,100])
mylabels=["carbon","hydrogen","oxygen"]
plt.pie(ny,labels=mylabels)
plt.show()
```

```python
#start angle
ny=np.array([20,34,80,120])
mylabels=["carbon","hydrogen","oxygen","nitrogen"]
plt.pie(ny,labels=mylabels,startangle=90)
plt.show()
```

```python
#explode
ny=np.array([20,34,80,120])
mylabels=["carbon","hydrogen","oxygen","nitrogen"]
myexplode=[0,0.25,0.95,0.8]
plt.pie(ny,labels=mylabels,startangle=90,explode=myexplode)
plt.show()
```

```python
#change colors
ny=np.array([20,34,80,120])
mylabels=["carbon","hydrogen","oxygen","nitrogen"]
mycolors=["darkblue","pink","black","yellow"]
plt.pie(ny,labels=mylabels,colors=mycolors,startangle=90)
plt.show()
```

```python
#legend
ny=np.array([20,34,80,120])
mylabels=["carbon","hydrogen","oxygen","nitrogen"]
plt.pie(ny)
plt.legend(labels=mylabels)
plt.show()
```
